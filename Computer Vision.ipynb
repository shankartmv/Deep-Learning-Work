```python
import cv2
import random
import numpy as np
```

```python
my_image=cv2.imread('F:\Study Guides\DeepLearning\Computer Vision\objects_for_contour.jpg')
cv2.namedWindow('Image for contour detection',cv2.WINDOW_GUI_EXPANDED)
cv2.imshow('Image for contour detection',my_image)
cv2.waitKey(0)
cv2.destroyWindow('Image for contour detection')
```

```python
gray=cv2.cvtColor(my_image,cv2.COLOR_BGR2GRAY)
edged=cv2.Canny(gray,50,120)
#ret_val,inverted=cv2.threshold(edged.copy(),30,255,cv2.THRESH_BINARY_INV)
contours,hierarchy=cv2.findContours(gray,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
#cv2.namedWindow('Canny Edges',cv2.WINDOW_GUI_EXPANDED)
#cv2.imshow('Canny Edges',edged)
#cv2.waitKey(0)
cv2.drawContours(my_image,contours,-1,(0,255,0),1)
cv2.namedWindow('Contours',cv2.WINDOW_GUI_EXPANDED)
cv2.imshow('Contours',my_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hindi_sentence=cv2.imread('F:\Study Guides\DeepLearning\Computer Vision\character segregation\hindi_sentence.jpg',cv2.IMREAD_GRAYSCALE)
row,col=hindi_sentence.shape
```

```python
print(hindi_sentence.shape)
cv2.namedWindow('Hindi',cv2.WINDOW_NORMAL)
cv2.imshow('Hindi',hindi_sentence)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blurred=cv2.GaussianBlur(hindi_sentence,(9,9),0)
#cv2.imshow('Blurred Image',blurred)
blur_edges=cv2.Canny(blurred,30,150)
#cv2.imshow('Detected Edges with blurring',blur_edges)
no_blur_edges=cv2.Canny(hindi_sentence,30,150)
#cv2.imshow('Detected Edges without blurring',no_blur_edges)
contour,hierarchy=cv2.findContours(blur_edges,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
print(len(contour))
background_black_image=np.zeros(shape=(row,col,3))
for indx,cnt in enumerate(contour):
    cont_image=cv2.drawContours(background_black_image,contour,indx,(0,10,0),0,hierarchy=hierarchy[0],maxLevel=0)
    cv2.imshow('identified_contours',cont_image)
    cv2.waitKey(0)
cv2.destroyAllWindows()
```
